#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).

# Review the spread return DataFrame
# YOUR CODE HERE

spread_return_early = arbitrage_spread_early[arbitrage_spread_early > 0] / bitstamp.loc['2018-02-01']
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0] / bitstamp.loc['2018-02-24']
spread_return_late = arbitrage_spread_late[arbitrage_spread_late > 0] / bitstamp.loc['2018-03-26']

print(spread_return_early)
print(spread_return_middle)
print(spread_return_late)
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs

# Review the first five profitable trades
# YOUR CODE HERE

# Define the minimum threshold for profitable trades

threshold = .1
spread_return_early = arbitrage_spread_early[arbitrage_spread_early > 0] / bitstamp.loc['2018-02-01']
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0] / bitstamp.loc['2018-02-24']
spread_return_late = arbitrage_spread_late[arbitrage_spread_late > 0] / bitstamp.loc['2018-03-26']

# Calculate the number of times the spread returns exceed the threshold for each date
profitable_trades_early = spread_return_early[spread_return_early > threshold].dropna().count()
profitable_trades_middle = spread_return_middle[spread_return_middle > threshold].dropna().count()
profitable_trades_late = spread_return_late[spread_return_late > threshold].dropna().count()

# Print the results
print("Number of profitable trades (early):", profitable_trades_early)
print("Number of profitable trades (middle):", profitable_trades_middle)
print("Number of profitable trades (late):", profitable_trades_late)
```
